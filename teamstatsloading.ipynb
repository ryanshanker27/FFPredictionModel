```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
```

```python
teamstats = pd.DataFrame()
for i in range(2004, 2023):
    x = pd.read_html('https://www.teamrankings.com/nfl/stat/points-per-game?date=%i-03-15' % (i+1))
    df = x[0]
    df['Year'] = df.columns[2]
    df = df[['Team', str(i), 'Year']]
    df = df.rename(columns = {str(i): 'TeamPPG'})
    x = pd.read_html('https://www.teamrankings.com/nfl/stat/opponent-points-per-game?date=%i-03-15' % (i+1))
    df2 = x[0]
    df2['Year'] = df2.columns[2]
    df2 = df2[['Team', str(i), 'Year']]
    df2 = df2.rename(columns = {str(i): 'teamOPPG'})
    df = pd.merge(df, df2, on = ['Team', 'Year'])
    x = pd.read_html('https://www.teamrankings.com/nfl/stat/yards-per-game?date=%i-03-15' % (i+1))
    df2 = x[0]
    df2['Year'] = df2.columns[2]
    df2 = df2[['Team', str(i), 'Year']]
    df2 = df2.rename(columns = {str(i): 'TeamYPG'})
    df = pd.merge(df, df2, on = ['Team', 'Year'])
    x = pd.read_html('https://www.teamrankings.com/nfl/stat/passing-play-pct?date=%i-03-15' % (i+1))
    df2 = x[0]
    df2['Year'] = df2.columns[2]
    df2 = df2[['Team', str(i), 'Year']]
    df2 = df2.rename(columns = {str(i): 'TeamPassPct'})
    df2['TeamPassPct'] = df2['TeamPassPct'].replace('%', '', regex = True)
    df = pd.merge(df, df2, on = ['Team', 'Year'])
    teamstats = pd.concat([teamstats, df])
```

```python
teamstats['Year'] = teamstats['Year'].astype(int)
teamstats['TeamPassPct'] = teamstats['TeamPassPct'].astype(float)
```

```python
teamstats = teamstats.reset_index().drop(columns = 'index')
```
