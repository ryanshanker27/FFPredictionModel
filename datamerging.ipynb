```python
finaltes = pd.read_csv('finaltes.csv')
finalrbs = pd.read_csv('finalrbs.csv')
finalwrs = pd.read_csv('finalwrs.csv')
finalqbs = pd.read_csv('finalqbs.csv')
```

```python
teamstats = pd.read_csv('teamstats.csv')
```

```python
teamstats['Year'] = teamstats['Year'] + 1
```

```python
teamdict = dict({'CAR': 'Carolina',
                'ARI': 'Arizona',
                'WAS': 'Washington',
                'CIN': 'Cincinnati',
                'TAM': 'Tampa Bay',
                'STL': 'LA Rams',
                'MIA': 'Miami',
                'IND': 'Indianapolis',
                'PIT': 'Pittsburgh',
                'NYG': 'NY Giants',
                'DAL': 'Dallas',
                'GNB': 'Green Bay',
                'OAK': 'Las Vegas',
                'DEN': 'Denver',
                'SDG': 'LA Chargers',
                'KAN': 'Kansas City',
                'JAX': 'Jacksonville',
                'NOR': 'New Orleans',
                'NWE': 'New England',
                'SEA': 'Seattle',
                'BAL': 'Baltimore',
                'CLE': 'Cleveland',
                'BUF': 'Buffalo',
                'DET': 'Detroit',
                'NYJ': 'NY Jets',
                'PHI': 'Philadelphia',
                'SFO': 'San Francisco',
                'CHI': 'Chicago',
                'TEN': 'Tennessee',
                'MIN': 'Minnesota',
                'HOU': 'Houston',
                'ATL': 'Atlanta',
                'LAR': 'LA Rams',
                'LAC': 'LA Chargers',
                'LVR': 'Las Vegas'})
```

```python
finaltes['Tm'] = finaltes['Tm'].replace(teamdict)
finalwrs['Tm'] = finalwrs['Tm'].replace(teamdict)
finalrbs['Tm'] = finalrbs['Tm'].replace(teamdict)
finalqbs['Tm'] = finalqbs['Tm'].replace(teamdict)
```

```python
finalwrs[finalwrs['Tm'] == '3TM']
```

```python
teamstats = teamstats.rename(columns = {'Team': 'Tm'})
```

```python
teamstats.columns
```

```python
finalwrs = pd.merge(finalwrs, teamstats[['Tm', 'TeamPPG', 'Year', 'teamOPPG', 'TeamYPG',
       'TeamPassPct']], on = ['Tm', 'Year'], how = 'left')
finalrbs = pd.merge(finalrbs, teamstats[['Tm', 'TeamPPG', 'Year', 'teamOPPG', 'TeamYPG',
       'TeamPassPct']], on = ['Tm', 'Year'], how = 'left')
finalqbs = pd.merge(finalqbs, teamstats[['Tm', 'TeamPPG', 'Year', 'teamOPPG', 'TeamYPG',
       'TeamPassPct']], on = ['Tm', 'Year'], how = 'left')
finaltes = pd.merge(finaltes, teamstats[['Tm', 'TeamPPG', 'Year', 'teamOPPG', 'TeamYPG',
       'TeamPassPct']], on = ['Tm', 'Year'], how = 'left')
```

```python
finalwrs.to_csv('finalwrs_wteams.csv')
finalrbs.to_csv('finalrbs_wteams.csv')
finalqbs.to_csv('finalqbs_wteams.csv')
finaltes.to_csv('finaltes_wteams.csv')
```
